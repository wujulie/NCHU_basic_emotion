```
import cv2
from pylab import *
import os
import time
import numpy as np
        
path='/content/drive/MyDrive/basic_emotion/cut_result'
filename='JAFFE_gray'
folders, x, y= [], [], []
for root, dirs, files in os.walk(path, topdown=False):
    for name in dirs:
        folders.append(name)
folders = folders[-7:]
start_time = time.time()
count = 0
for fld in folders:
    index = folders.index(fld)
    for root, dirs, files in os.walk(path + '/' + fld, topdown=False):
        for file in files:
            filepath = root + '/' + file
            img = cv2.imread(filepath)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            resized = cv2.resize(img, (224, 224), cv2.INTER_LANCZOS4)
            y.append(index)
            x.append(resized)
    print(fld)

np.save('{}_X.npy'.format(filename), x)
np.save('{}_Y.npy'.format(filename), y)
print("Complete")

```
